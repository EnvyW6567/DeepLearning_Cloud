```python
import numpy as np

x = np.array([0.5, 0.8, 0.2])
w = np.array([0.4, 0.7, 0.8])
d = 1
e = []
e.append(d)
alpha = 0.5

def SIGMOID(x):
    return 1 / (1 + np.exp(-x))

for i in range(100):
    v = np.sum(w * x)
    y = SIGMOID(v)
    e.append(d - y)
    print(f'error {i+1:2} : {e[i+1]:0.6f} / {(e[i]-e[i+1])/e[i]*100:0.6f}%')
    w = w + alpha * y * (1-y) * e[i+1] * x

```
